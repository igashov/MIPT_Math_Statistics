```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

%matplotlib inline
```

```python
# Работаем с выборкой размера N = 10000 из стандартного нормального распределения.
N = 10000
```

```python
# Генерируем выборку.
sample = norm.rvs(size=N)

plt.figure(figsize=(10,1))
plt.scatter(sample, np.zeros(N), alpha=0.2)
plt.show()
```

```python
# Эмпирическая функция распределения
def empirical_cdf(n, x, sample):
    return sum(item <= x for item in sample[:n]) / n
```

```python
# Для n = 10, 25, 50, 1000, 10000 строим графики эмпирической ф.р., нанося на каждый из них также истинную ф.р.
for n in [10, 25, 50, 1000, N]:
    grid = np.linspace(-4, 4, 1000)
    emp_cdf = np.array([empirical_cdf(n, x, sample) for x in grid])
    
    plt.figure(figsize=(10, 5))
    plt.plot(grid, emp_cdf, label='empirical CDF')
    plt.scatter(sample[:n], np.zeros(n) - 0.1, alpha=0.2, label='sample')
    plt.plot(grid, norm.cdf(grid), color='red', label='CDF')
    # Для обозначения скачков воспользуемся вертикальными линиями погрешностей.
    plt.errorbar(sample[:n], np.zeros(n) - 1, yerr=2*n, alpha=0.1, barsabove=True, color='grey') 
        
    plt.ylim(-0.15, 1.1)
    plt.xlim(-5, 5)
    plt.title(r'CDF and empirical CDF for ' + str(n) + ' variables.')
    plt.legend(fontsize=10, loc=2)
    plt.show()
```

```python
# Статистики 
variance = np.zeros(N)
for n in range(1, N, 1):
    emp_cdf = np.array([empirical_cdf(n, x, sample) for x in grid])
    variance[n] = max(abs(emp_cdf - norm.cdf(grid)))

n_variance = variance * [n ** 0.5 for n in range(1, N + 1, 1)]
```

```python
len(variance)
```

```python
# Графики
plt.figure(figsize=(10, 5))
plt.plot(np.linspace(1, N, N), variance, label='$D_n(X)$')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(np.linspace(1, N, N), n_variance, label='$\sqrt{n}D_n(X)$')
plt.show()
```
