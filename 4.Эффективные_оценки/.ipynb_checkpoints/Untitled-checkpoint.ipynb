```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
from scipy.stats import expon
from scipy.stats import norm
from scipy.stats import beta

%matplotlib inline
```

```python
N = 1000
m = 50
sigma2 = 2.1
theta = beta.rvs(1, 1, size=1)

sample_binom = binom.rvs(m, theta, size=N)
sample_expon = expon.rvs(scale=theta, size=N)
sample_norm = norm.rvs(theta, sigma2, size=N)
```

```python
# Эффективные оценки
eff_binom = sample_binom.cumsum() * [1 / (n + 1) for n in range(N)] / m
eff_expon = sample_expon.cumsum() * [1 / (n + 1) for n in range(N)]
eff_norm = sample_norm.cumsum() * [1 / (n + 1) for n in range(N)]
```

```python
# Для каждого распределения построили по 500 бутстрепных выборок
K = 500

bootstrap_samples_binom = [binom.rvs(m, eff_binom[-1], size=N) for k in range(K)]
bootstrap_samples_expon = [expon.rvs(scale=eff_expon[-1], size=N) for k in range(K)]
bootstrap_samples_norm = [norm.rvs(eff_norm[-1], sigma2, size=N) for k in range(K)]
```

```python
# Для каждой бутстрепной выборки посчитали эффективную оценку для каждого n = 1,2,...,N
bootstrap_estimates_binom = np.array([bootstrap_samples_binom[k].cumsum() * [1 / (n + 1) for n in range(N)] / m 
                             for k in range(K)])
bootstrap_estimates_expon = np.array([bootstrap_samples_expon[k].cumsum() * [1 / (n + 1) for n in range(N)] 
                             for k in range(K)])
bootstrap_estimates_norm = np.array([bootstrap_samples_norm[k].cumsum() * [1 / (n + 1) for n in range(N)] 
                            for k in range(K)])                       
```

```python
# Для каждого n считаем выборочную дисперсию по соответствующим оценкам
bootstrap_var_binom = np.array([np.sum(bootstrap_estimates_binom[:,n] ** 2) / K - 
                       (np.sum(bootstrap_estimates_binom[:,n]) / K) ** 2 for n in range(N)])
bootstrap_var_expon = np.array([np.sum(bootstrap_estimates_expon[:,n] ** 2) / K - 
                       (np.sum(bootstrap_estimates_expon[:,n]) / K) ** 2 for n in range(N)])
bootstrap_var_norm = np.array([np.sum(bootstrap_estimates_norm[:,n] ** 2) / K - 
                       (np.sum(bootstrap_estimates_norm[:,n]) / K) ** 2 for n in range(N)])
```

```python
# Посчитаем информацию Фишера для каждого n:
Fisher_binom = np.linspace(1, N, N) * m / (theta * (1 - theta))
Fisher_expon = np.linspace(1, N, N) / (theta ** 2)
Fisher_norm = np.linspace(1, N, N) / sigma2
```

```python
# Строим графики:
plt.figure(figsize=(10, 5))
plt.plot(np.linspace(1., N, N), bootstrap_var_binom, label=r'$Variance$')
plt.plot(np.linspace(1., N, N), 1  / Fisher_binom, label=r'$\frac{1}{I_X(\theta)}$')
plt.ylim(0, 0.00002)

plt.legend()
plt.title(r'Binom')
plt.xlabel(r'Size of sample', fontsize='10')
plt.grid()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(np.linspace(1., N, N), bootstrap_var_expon, label=r'$Variance$')
plt.plot(np.linspace(1., N, N), 1  / Fisher_expon, label=r'$\frac{1}{I_X(\theta)}$')
plt.ylim(0, 0.0002)

plt.legend()
plt.title(r'Expon')
plt.xlabel(r'Size of sample', fontsize='10')
plt.grid()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(np.linspace(1., N, N), bootstrap_var_norm, label=r'$Variance$')
plt.plot(np.linspace(1., N, N), 1  / Fisher_norm, label=r'$\frac{1}{I_X(\theta)}$')
plt.ylim(0, 0.2)

plt.legend()
plt.title(r'Norm')
plt.xlabel(r'Size of sample', fontsize='10')
plt.grid()
plt.show()
```
